```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
```

```python
a = pd.read_csv("/content/drive/MyDrive/ML Datasets/gold-price.csv")
df = pd.DataFrame(a)
```

```python
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Years"] = df["Date"].dt.year
```

```python
df["Months"] = df["Date"].dt.month
```

```python
df["Days"] = df["Date"].dt.day
```

```python
df = df.drop(columns="Date")
```

```python
df
```

```python
X = df.drop(columns="GLD").values
```

```python
Y = df["GLD"].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, shuffle=True)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
from sklearn import metrics
```

```python
X_train_rsq = metrics.r2_score(X_train_prediction, Y_train)
print(X_train_rsq)
```

```python
X_test_prediction = model.predict(X_test)
```

```python
X_test_rsq = metrics.r2_score(X_test_prediction, Y_test)
print(X_test_rsq)
```

```python
import gradio as gr
import numpy as np
import pandas as pd
import joblib  # or use pickle if needed

# Load your trained model and scaler
model = joblib.load("gold_model.pkl")     # Replace with your actual model file
scaler = joblib.load("gold_scaler.pkl")   # Replace with your actual scaler file

def predict_gold(date, spx, usd_index, silver, eurusd):
    try:
        # Parse the date
        parsed_date = pd.to_datetime(date, format="%d/%m/%Y")
        year = parsed_date.year
        month = parsed_date.month
        day = parsed_date.day

        # Combine inputs in correct order
        final_input = [spx, usd_index, silver, eurusd, day, month, year]

        # Convert to NumPy, reshape, and scale
        input_array = np.asarray(final_input).reshape(1, -1)
        input_scaled = scaler.transform(input_array)

        # Predict
        prediction = model.predict(input_scaled)
        return f"📈 Predicted Gold Price: {prediction[0]:.2f}"

    except Exception as e:
        return f"❌ Error: {str(e)}"

# Gradio Interface
interface = gr.Interface(
    fn=predict_gold,
    inputs=[
        gr.Textbox(label="Date (DD/MM/YYYY)"),
        gr.Number(label="S&P"),
        gr.Number(label="USD Index"),
        gr.Number(label="Silver"),
        gr.Number(label="EUR/USD"),
    ],
    outputs="text",
    title="Gold Price Prediction App",
    description="Enter date and financial indicators to predict the gold price.",
)

interface.launch()
```
